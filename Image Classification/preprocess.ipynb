```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import glob as gb 
import cv2
```

```python
train_path = '/home/vaibhav/Desktop/AI/Image Classification/seg_test'
for folder in os.listdir(train_path):
    images=gb.glob(pathname=str(train_path + '/' +folder +'/*.jpg'))
    print(f'for training data, found {len(images)} in folder {folder}')
```

```python
code={'buildings':0 ,'forest':1,'glacier':2,'mountain':3,'sea':4,'street':5}
def getname(n):

    for k,v in code.items():
        if v==n:
            return k
```

```python
x_train=[]
y_train=[]
for folder in os.listdir(train_path):
    images=gb.glob(pathname=str(train_path + '/'+ folder +'/*.jpg'))
    for img in images:
        img=cv2.imread(img)
        img=cv2.resize(img,(100,100))
        x_train.append(img)
        y_train.append(code[folder])
```

```python
plt.figure(figsize=(10,10))
for i,v in enumerate(np.random.randint(0,len(x_train),9)):
    plt.subplot(3,3,i+1)
    plt.imshow(x_train[v])
    plt.title(getname(y_train[v]))
    plt.axis('off')
```

```python
test_path  = '/home/vaibhav/Desktop/AI/Image Classification/seg_train'

x_test=[]
y_test=[]
for folder in os.listdir(test_path):
    images=gb.glob(pathname=str(test_path + '/' + folder +'/*.jpg'))
    for img in images:
        img=cv2.imread(img)
        img=cv2.resize(img,(100,100))
        x_test.append(img)
        y_test.append(code[folder])
```

```python
class_names = ['mountain', 'street', 'buildings', 'sea', 'forest', 'glacier']
class_names_label = {class_names:i for i,class_names in enumerate(class_names)}
class_names_label
```

```python
x_train=np.array(x_train)
x_train = x_train / 255.0
x_test=np.array(x_test)
x_test = x_test / 255.0
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
print(f'X_train shape  is {x_train.shape}')
print(f'X_test shape  is {x_test.shape}')
print(f'y_train shape  is {y_train.shape}')
print(f'y_test shape  is {y_test.shape}')
```

```python
y_train_1 = np.zeros((3000, 6))
for i in range(0, 3000):
    ind = y_train[i]
    one_hot = np.zeros((6, ))
    one_hot[ind] = 1
    print(ind)
    print(one_hot)
    y_train_1[i] = one_hot
```

```python
y_test_1 = np.zeros((14034, 6))
for i in range(0, 14034):
    ind = y_test[i]
    one_hot = np.zeros((6, ))
    one_hot[ind] = 1
    print(ind)
    print(one_hot)
    y_test_1[i] = one_hot
```
